```python
#What categories of applications get a lot of downloads per day?
grp=df.groupby(['genre','min_installs'])
list1=df['genre'].unique()
print(len(list1))
dict2={}

for group,value in grp:
    for i in range(len(list1)):
        if(group[0]==list1[i]):
            if group[0] not in dict2:
                dict2[group[0]]=group[1]
            else:
                dict2[group[0]]+=group[1]
print(dict2)
a = sorted(dict2.items(), key=lambda x: x[1])
print("Total downloads per day",a)
print("maximum lot of downloads per day ",a[-1])
a[-1][1]
            
    
    
    #print(group[1])
    #print((group[1]*value["min_installs"].count()))
    
```

```python
import numpy as np
import pandas as pd
from datetime import date
df['released'] = pd.to_datetime(df['released'])
today = date.today()
d1 = today.strftime("%Y-%m-%d")
today = np.datetime64(d1[:4]+'-'+d1[5:7]+'-'+d1[8:10])
lowest = df['released'].min()
latest = df['released'].max()
print(lowest,latest)
diff = today - lowest
print(int(diff.days))
```

```python
g2 = df.groupby(['genre'])

dict1 = g2.agg({'released': np.min})

d = dict1.to_dict()

min_date = {}
#days_since = {}

for values in d.values():
    for k,v in values.items():
        min_date[k] = v

dates = pd.DataFrame.from_dict(min_date,orient='index')

print(dates)
dates.dtypes
```
